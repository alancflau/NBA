## Grab information on player stats Start 1988

```python
import nba_stats
```

```python
df_per_game = nba_stats.player_stats(2019, 'per_game')
df_advanced = nba_stats.player_stats(2019, 'advanced')
```

```python
# Determine the matchin column names in these two dataframes
intersect_cols = df_per_game.columns.intersection(df_advanced.columns).tolist()
intersect_cols.remove('MP')

ex = df_per_game.merge(df_advanced, how = 'left', on = intersect_cols)
```

```python
combined_df = ex.dropna().reset_index(drop = True)
```

```python
# combined_df[combined_df['Player'].str.contains('Artest')]
```

```python
def player_stats_seasons(season_start, season_end):
    
    all_dfs = []
    for year in range(season_start, season_end+1):
        
        df_per_game = nba_stats.player_stats(year, 'per_game')
        df_advanced = nba_stats.player_stats(year, 'advanced')
        
        intersect_cols = df_per_game.columns.intersection(df_advanced.columns).tolist()
        intersect_cols.remove('MP')

        ex = df_per_game.merge(df_advanced, how = 'left', on = intersect_cols)
        ex['Season'] = year       
        combined_df = ex.dropna().reset_index(drop = True)

        all_dfs.append(combined_df)
    
        result = pd.concat(all_dfs)
    
    return result
```

```python
result = player_stats_seasons(1989,2019)
```

Create a dictionary of years abbreviated and write the full year

```python
mydict = {}

for i in range(89,100,1):
    mydict[str(i)] = '19{}'.format(i)
    
for i in range(0,21,1):
    
    if len(str(i)) == 1:
        mydict[('%0.2d' % i)] = '200{}'.format(i)
    else:   
        mydict[str(i)] = '20{}'.format(i)
```
